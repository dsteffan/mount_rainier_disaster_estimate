```python
def display_shp(path, color = "r"):
    shp = pysal.lib.io.open(path)
    fig, ax = matplotlib.pyplot.subplots(figsize=(10,6))
    for state in range(len(list(shp))):
        x = []
        y = []
        for i in shp[state].vertices:
            x.append(i[0])
            y.append(i[1])

        matplotlib.pyplot.scatter(x, y, c = color, s = 1)
        matplotlib.pyplot.xlim(shp.header['BBOX Xmin'], shp.header['BBOX Xmax'])
        matplotlib.pyplot.ylim(shp.header['BBOX Ymin'], shp.header['BBOX Ymax'])
```

```python
display_shp('../data/rainier_98shapefiles/case1.shp', color = "b")
```
